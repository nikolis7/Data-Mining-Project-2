The data are from a study performed in 1952 in United States and present per county the cooperation of 114 religious denominations.

The attributes of the dataset are the following:

* **Case ID** : A distict ID for every instance 
* **CNAME**: The name of county in conjunction with state two digit ISO separated with a comma
* **TOTPOP** : The total population of the county 
* **TOTCHUR** : The total churches in the county 
* **_M**: There are 114 attributes which end with _M for every different religion and contains the number of the members 
* **_C** : There are 114 attributes which end with _C for every different religion and contains the number of the churches

### Can you find the 3 most extreme counties with respect to the distribution of their churches across religions?

```python
from sklearn.ensemble import IsolationForest # Multivalue outlier detection on religious groups distribution
from sklearn.cluster import DBSCAN

# We need to keep the data only with the counties and church distribution per religion 
df_cd = df.loc[:,c_C].drop(['STATE'], axis=1)

df_cd.set_index('COUNTY', inplace=True)


random_state = np.random.RandomState(42)

# initialize model 
model = IsolationForest(n_estimators=100,max_samples='auto',contamination=float(0.2),random_state=random_state)


#fit the model 
model.fit(df_cd)


# assign outlier score into data 
df_cd['scores'] = model.decision_function(df_cd)

df_cd['anomaly_score'] = model.predict(df_cd.drop('scores',axis=1))


# sort the values based on scores 

df_cd.sort_values(by='scores' , ascending=True, inplace=True) # ascending because we want the lower score
```

The anomaly score of the input samples. The lower, the more abnormal

```python
# print the top three anomalies, which indicate extreme distribution regarding churches    
df_cd[df_cd['anomaly_score']==-1].head(3)
```

From the previous analysis we identify that there are many counties without churches for most of the religions (df_cd is a sparse table). Thus, the outlier detection algorithm used indicate as outlier the counties with the existence of many religion churches.
